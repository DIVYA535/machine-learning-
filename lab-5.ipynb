## Handling Missing Values

```python
# Replace missing values in 'Embarked' column with 'S'
df["Embarked"].fillna("S", inplace=True)
# Verify if missing values are replaced
print(df["Embarked"].isnull().sum())  # Should print 0
```

```python
# Replace missing values in 'Fare' column with its median value
df["Fare"].fillna(df["Fare"].median(), inplace=True)

# Display the first few rows to confirm the changes
print(df.head())

# Check if missing values are replaced
print("\nMissing values in 'Fare':", df["Fare"].isnull().sum())  # Should print 0
```

```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("train.csv")

# Compute mean and standard deviation of existing Age values
mean_age = df["Age"].mean()
std_age = df["Age"].std()

# Count missing values in Age
num_missing = df["Age"].isnull().sum()

# Generate random ages from the same distribution
random_ages = np.random.normal(loc=mean_age, scale=std_age, size=num_missing)

# Fill missing Age values with these random numbers
df.loc[df["Age"].isnull(), "Age"] = random_ages

# Display the first few rows to confirm changes
print(df.head())

# Check if missing values are replaced
print("\nMissing values in 'Age':", df["Age"].isnull().sum())  # Should print 0
```

```python
# Label encoding of categorical features

import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load the dataset
df = pd.read_csv("train.csv")

# Drop unnecessary columns
df.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"], inplace=True)

# Identify categorical features
categorical_features = ["Sex", "Embarked"]

# Apply Label Encoding
label_encoders = {}  # Dictionary to store encoders for each feature
for feature in categorical_features:
    le = LabelEncoder()
    df[feature] = le.fit_transform(df[feature].astype(str))  # Convert to string and encode
    label_encoders[feature] = le  # Save the encoder for future use

# Display first few rows after encoding
print(df.head())

# Check encoded values
for feature, encoder in label_encoders.items():
    print(f"\nEncoding for {feature}: {dict(zip(encoder.classes_, encoder.transform(encoder.classes_)))}")

```

```python
# Create the model
# Fitting the decision tree with default hyperparameters
# Importing decision tree classifier from sklearn library


import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

df = pd.read_csv("train.csv").drop(columns=["PassengerId", "Name", "Ticket", "Cabin"])
df.fillna(df.median(numeric_only=True), inplace=True)  
df["Sex"], df["Embarked"] = df["Sex"].astype("category").cat.codes, df["Embarked"].astype("category").cat.codes  

X, y = df.drop("Survived", axis=1), df["Survived"]
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.1, random_state=42)

print("Accuracy:", DecisionTreeClassifier(random_state=42).fit(X_train, y_train).score(X_val, y_val))
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

clf = DecisionTreeClassifier().fit(X_train, y_train)  

plt.figure(figsize=(15, 10))  # Adjust width & height
plot_tree(clf, feature_names=X_train.columns, class_names=["Not Survived", "Survived"], filled=True)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Load and preprocess data
data = pd.read_csv('train.csv').apply(LabelEncoder().fit_transform)

# Fit the model and plot the tree
X, y = data.iloc[:, :-1], data.iloc[:, -1]
model = DecisionTreeClassifier(max_depth=3).fit(X, y)
plot_tree(model, filled=True)
```
