## Handling Missing Values

# Replace the msiing values in Embarked column by 'S'


```python
# Replace missing values in the 'Embarked' column with 'S'
train_df['Embarked'].fillna('S', inplace=True)

# Verify if there are any missing values left in 'Embarked'
print(train_df['Embarked'].isnull().sum())  # Should print 0
```

# Impute the missing values in Fare column with median value


```python
# Replace missing values in the 'Fare' column with the median value
train_df['Fare'].fillna(train_df['Fare'].median(), inplace=True)

# Verify if there are any missing values left in 'Fare'
print(train_df['Fare'].isnull().sum())  # Should print 0
```

# Impute the Age column. It has too many missing values but it may be important. 
# We fill the missing values with numbers randomly drawn from distribution
# that has same mean and standard deviation as non missing Age data

```python
import numpy as np

# Calculate the mean and standard deviation of the existing Age values
age_mean = train_df['Age'].mean()
age_std = train_df['Age'].std()

# Count missing Age values
missing_count = train_df['Age'].isnull().sum()

# Generate random values from a normal distribution with same mean & std
random_ages = np.random.normal(loc=age_mean, scale=age_std, size=missing_count)

# Ensure all generated ages are non-negative
random_ages = np.clip(random_ages, 0, None)

# Fill missing Age values with these randomly generated numbers
train_df.loc[train_df['Age'].isnull(), 'Age'] = random_ages

# Check if all missing values are filled
print("Missing Age values after imputation:", train_df['Age'].isnull().sum())  # Should print 0
```

# Print the tree

# Importing required packages for visualization


```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt

# Load the dataset
train_df = pd.read_csv('train.csv')

# Preprocessing: Label encode categorical columns and handle missing values
train_df = train_df.apply(LabelEncoder().fit_transform)  # Encode all categorical columns
X = train_df.drop(columns=['Survived', 'PassengerId', 'Name', 'Ticket', 'Cabin'])
y = train_df['Survived']

# Split into train and validation sets
X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size=0.1, random_state=42)

# Impute missing values
imputer = SimpleImputer(strategy='mean')
X_train = imputer.fit_transform(X_train)
X_validation = imputer.transform(X_validation)

# Train Decision Tree model
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

# Plot the tree
plt.figure(figsize=(20, 10))
plot_tree(model, feature_names=X.columns, class_names=['Not Survived', 'Survived'], filled=True, rounded=True)
plt.show()
```

# plotting tree with max_depth=3

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('train.csv')

# Encode categorical features
for col in ['Sex', 'Embarked']:
    df[col] = LabelEncoder().fit_transform(df[col].astype(str))

# Select features and target
X = df.drop(columns=['Survived', 'PassengerId', 'Name', 'Ticket', 'Cabin'])
y = df['Survived']

# Train-test split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.1, random_state=42)

# Handle missing values
X_train = SimpleImputer(strategy='mean').fit_transform(X_train)
X_val = SimpleImputer(strategy='mean').fit_transform(X_val)

# Train decision tree
model = DecisionTreeClassifier(max_depth=3, random_state=42)
model.fit(X_train, y_train)

# Plot tree
plt.figure(figsize=(15, 8))
plot_tree(model, feature_names=X.columns, class_names=['Not Survived', 'Survived'], filled=True, rounded=True)
plt.show()
```
